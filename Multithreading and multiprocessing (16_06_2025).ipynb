# MulitiThreading

```python
import threading
import time

def task():
  print("doing a small task")
  time.sleep(1)  #makes the thread sleep for i second
  print("task is done")

#create and start two threads

t1 = threading.Thread(target=task)
t2 = threading.Thread(target=task)

t1.start()
t2.start()

t1.join() # wait here until t1 finishes
t2.join() # wait here until t2 finishes

print("done")
```

```python
import threading
import time

def boil_water():
  print(f"[{time.strftime('%x')}] boiling the water")
  time.sleep(3)  #makes the thread sleep for 3 second
  print(f"[{time.strftime('%x')}] done boiling")

def chop_vegies():
  print(f"[{time.strftime('%x')}] chopping veggies")
  time.sleep(2)  #makes the thread sleep for 2 second
  print(f"[{time.strftime('%x')}] done chopping")

#create and start two threads
print(f"[{time.strftime('%x')}] starting tasks")

t1 = threading.Thread(target=boil_water)
t2 = threading.Thread(target=chop_vegies)

t1.start()
t2.start()

t1.join() # wait here until t1 finishes
t2.join() # wait here until t2 finishes

print(f"[{time.strftime('%x')}] all tasks done")
```

# MultiProcessing

```python
import multiprocessing
import time

def boilrice():
  print(f"[{time.strftime('%H:%M:%S')}] boiling rice...")
  time.sleep(5)
  print(f"[{time.strftime('%H:%M:%S')}] rice is ready")

def fryvegies():
  print(f"[{time.strftime('%H:%M:%S')}] frying veggies...")
  time.sleep(4)
  print(f"[{time.strftime('%H:%M:%S')}] veggies are ready")

def bakecake():
  print(f"[{time.strftime('%H:%M:%S')}] baking a cake...")
  time.sleep(6)
  print(f"[{time.strftime('%H:%M:%S')}] cake is ready")

if __name__ == '__main__':
  print(f"[{time.strftime('%H:%M:%S')}] ------ starting tasks ------")

  #create indivisual process
  p1 = multiprocessing.Process(target=boilrice)
  p2 = multiprocessing.Process(target=fryvegies)
  p3 = multiprocessing.Process(target=bakecake)

  #start all the process
  p1.start()
  p2.start()
  p3.start()

  print(f"[{time.strftime('%H:%M:%S')}] ------ processing... ------")

  #wait for all the process to finish
  p1.join()
  p2.join()
  p3.join()

  print(f"[{time.strftime('%H:%M:%S')}] ------  all tasks done ------")
```

QI .Write two functions: boil_milk() and toast_bread() that simulate boiling milk (3s) and toasting bread (2s). Use two
threads to run them simultaneously.

```python
import threading
import time

def boil_milk():
  print(f"[{time.strftime('%H:%M:%S')}] boiling milk...")
  time.sleep(3)
  print(f"[{time.strftime('%H:%M:%S')}] milk is ready")

def toast_bread():
  print(f"[{time.strftime('%H:%M:%S')}] toasting bread...")
  time.sleep(2)
  print(f"[{time.strftime('%H:%M:%S')}] toast ready")

if __name__ == '__main__':
  print(f"[{time.strftime('%x')}] ------ starting tasks ------")

  #create indivisual process
  p1 = threading.Thread(target=boil_milk)
  p2 = threading.Thread(target=toast_bread)

  #start all the process
  p1.start()
  p2.start()

  print(f"[{time.strftime('%H:%M:%S')}] ------ processing... ------")

  #wait for all the process to finish
  p1.join()
  p2.join()

  print(f"[{time.strftime('%x')}] ------  all tasks done ------")
```
